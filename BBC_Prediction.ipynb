```python

# -----------------------------------
# BBC Sport Classification - Prediction Script
# -----------------------------------

import pickle

print("Loading model and vectorizer...")

# ABSOLUTE PATHS (guaranteed to work)
VECTORIZER_PATH = r"C:\Users\saad\BBCSPORT\bbcsport\models\bbc_vectorizer.pkl"
MODEL_PATH      = r"C:\Users\saad\BBCSPORT\bbcsport\models\bbc_svm_model.pkl"

# Load vectorizer
with open(VECTORIZER_PATH, "rb") as f:
    vectorizer = pickle.load(f)

# Load model
with open(MODEL_PATH, "rb") as f:
    model = pickle.load(f)

print("Model & vectorizer loaded successfully!")
print("-" * 50)

```

```python
# -----------------------------------
# Prediction function
# -----------------------------------
def predict_category(text: str):
    """Predict the BBC Sport category for a given news article text."""
    X = vectorizer.transform([text])
    pred = model.predict(X)[0]
    return pred


```

```python
# -----------------------------------
# Example prediction
# -----------------------------------
sample_text = """Roger Federer wins another major title after defeating Nadal in a thrilling match."""

prediction = predict_category(sample_text)

print("Input text:")
print(sample_text)
print("\nPredicted category:", prediction)
```
